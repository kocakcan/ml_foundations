## Dimensionality Reduction
A technique to reduce the number of features (dimensions) in a dataset while preserving as much information as possible

**Why is it needed?**
- As dimensions increase, data becomes sparse
- More data needed for statistical significance
- Computational complexity increases
- Model performance can degrade
- **Storage:** More memory needed
- **Computation:** Slower processing
- **Visualization:** Can't visualize > 3D
- **Noise:** More features can mean more noise

```python
# Example of high-dimensional data
import numpy as np

# 1000 samples with 100 features
X = np.random.randn(1000, 100)
```

**PCA Steps**
1) Standardize the data (zero mean, unit variance)
2) Compute covariance matrix
3) Calculate eigenvectors and eigenvalues
4) Sort eigenvectors by eigenvalues (descending)
5) Choose top k eigenvectors
6) Project data onto new subspace

## Main Applications of PCA
**Feature Selection/Extraction**
- Remove redundant features
- Combine correlated features
- Extract meaningful patterns

**Data Compression**
- Reduce storage requirements
- Speed up training
- Improve model efficiency

```python
# Data Visualization
from sklearn.decomposition import PCA

# Reduce 100D to 2D for visualization
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Now can plot in 2D
import matplotlib.pyplot as plt
plt.scatter(X_2d[:, 0], X_2d[:, 1])
plt.title('High-dimensional data in 2D')
plt.show()
```

**Benefits in ML**
- Reduces overfitting
- Speeds up training
- Removes multicollinearity
- Makes visualization possible
- Reduce noise in data

**When to Use**
- High-dimensional datasets
- Correlated features
- Need for visualization
- Memory/computational constraints
- Model performance issues

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Create high-dimensional dataset
X, y = make_classification(n_samples=1000, n_features=100, random_state=42)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Without dimensionality reduction
model1 = LogisticRegression()
model1.fit(X_train, y_train)
score1 = model1.score(X_test, y_test)

# With PCA
pca = PCA(n_components=20)  # Reduce to 20 dimensions
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)

model2 = LogisticRegression()
model2.fit(X_train_reduced, y_train)
score2 = model2.score(X_test_reduced, y_test)

print(f"Score without reduction: {score1:.3f}")
print(f"Score with reduction: {score2:.3f}")
print(f"Training time reduced, memory usage reduced")
```
