## The percentile
- The value below which a certain amount of data falls.
- If value X is at the kth percentile, then X is greater than K% of the values
- **The median is a special case of percentile** (the second percentile, Q2, or 50th percentile) and it splits the data in half

```python
import numpy as np
```

```python
sat_results = [610,1060,1270,675,820,940,900,1340,1420,625,850,1210,910,1200,1120,870,600,1135,1160,1050,960,500,1200,1000,800,1137,810,1230,680,1130,602,1015,1010,1170,1310,670,1420,1005,1135,1600,1540]
# finding the 90 percentile
sat_percentile = np.percentile(sat_results, 90)
print(f"90th percentile: {sat_percentile}")
# finding the 70 percentile
seventy_percentile = np.percentile(sat_results, 70)
print(f"70th percentile: {seventy_percentile}")
```
